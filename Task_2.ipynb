```python
import cv2
import numpy as np
#import time

cap = cv2.VideoCapture('ball.mp4')

#start = time.time()
y_initial = 0

if (cap.isOpened()== False):
    print("Error opening video stream or file")

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        
        black = np.zeros((650,650,3), np.uint8)
        
        grey = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        canny = cv2.Canny(grey,30,180)
        
        
        cop = canny.copy()
        contours, hierarchy = cv2.findContours(cop, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        
        
        for cnt in contours:
            area = cv2.contourArea(cnt)
            
            if area>3000:
                
                
                
                cent = cv2.moments(cnt)
                
                
                X = int(cent["m10"]/cent["m00"])
                Y = int(cent["m01"]/cent["m00"])
                
               # end = time.time()
               # duration = end - start
                diff = Y - y_initial
              #  velocity = diff/duration
                y_initial = Y
              #  start = end
               
                cv2.circle(black, (X,Y), 8, (255,255,255), -1)
                cv2.putText(black, "("+str(X)+","+str(Y)+")", (X-20, Y-20), cv2.FONT_HERSHEY_COMPLEX_SMALL,0.5,(220,0,0), 1)
                
                if diff>0:
                    cv2.arrowedLine(black, ((X-30), (Y+20)), ((X-30), (Y-20)), (0,0,225), 4)
                elif diff<0:
                    cv2.arrowedLine(black, ((X-30), (Y-20)), ((X-30), (Y+20)), (0,0,225), 4)
                else:
                    continue
                
        cv2.imshow('Contours',black)
        
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
            
        
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
